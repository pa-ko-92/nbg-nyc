# Doing a tiny economic analysis using python 


## The workflow is:
- find a website that contains interesting data
- scrape that website, i.e. extract the data
- write the data to a csv-file so you don't have to rescrape (most websites don't like that)
- generate a pandas DataFrame from the csv file
- use the DF for analysis

### importing all the required python modules

```python
import pandas as pd
import seaborn as sns
from bs4 import BeautifulSoup as bs4
import requests
print('importing done')
```

### downloading & scraping the page

```python
#getting data from: 
#https://www.worlddata.info/cost-of-living.php
url = "https://www.worlddata.info/cost-of-living.php"
req = requests.get(url)
soup = bs4(req.text, "html.parser")
print(soup.title.text,  '- downloaded successfully!')
```

```python
table = soup.find( "table", {"id":"tabsort"} )

allrows = []
for row in table.findAll("tr"):
    new_row = []
    for td in row.findAll("td"):
        new_row.append(td.text)
        
    allrows.append(new_row)

# first row is empty
#creating new first row with appropriate column names
allrows[0] = ['rank', 'country', 'cost index', 'avg monthly income (USD)', 'purchasing power index']
```

### create the csv file

```python
#writing content of rows in a csv-file
with open('cost_of_living_raw.csv', 'w') as myfile:
    for row in allrows:
        for elem in row:
            myfile.write(elem)
            #using ';' as delimiter, because '.' and ',' are used in the values
            myfile.write(';')
        myfile.write('\n')
    
```

### create the pandas DataFrame, sanitize the data

```python
costOfLiving = pd.read_csv('cost_of_living_raw.csv', index_col='rank', sep=';')
#drop last column 
costOfLiving.drop('Unnamed: 5', axis=1, inplace=True) 

#remove string 'USD' in avg monthly income
#it's easier to work with numeric values than with strings in pandas
#also remove ',' as 1000-seperator 

costOfLiving['avg monthly income (USD)'] = costOfLiving['avg monthly income (USD)'].map(lambda x: x.rstrip(' USD'))
costOfLiving['avg monthly income (USD)'] = costOfLiving['avg monthly income (USD)'].map(lambda x: x.replace(',', ''))
costOfLiving['avg monthly income (USD)'] = pd.to_numeric(costOfLiving['avg monthly income (USD)'])
costOfLiving.head(10)
```

### get a first overview of all cost-of-living values 

```python
sns.set(rc={'figure.figsize':(80,40)})
sns.barplot(x=costOfLiving['cost index'], y=costOfLiving.country)
```

### compare the cost of living in the USA & Germany and their southern and northern neighbors 

```python
list_of_countries = ['Germany', 'United States', 'Canada', 'Austria', 'Denmark', 'Mexico']
costIndex = costOfLiving[costOfLiving['country'].isin(list_of_countries)]['cost index']
countries = costOfLiving[costOfLiving['country'].isin(list_of_countries)]['country']

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x=countries, y=costIndex)
```

### compare the average monthly income in the USA & Germany and their southern and northern neighbors 

```python
list_of_countries = ['Germany', 'United States', 'Canada', 'Austria', 'Denmark', 'Mexico']
avgIncome = costOfLiving[costOfLiving['country'].isin(list_of_countries)]['avg monthly income (USD)']
countries = costOfLiving[costOfLiving['country'].isin(list_of_countries)]['country']

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x=countries, y=avgIncome)
```
